### Imagerie par R√©sonance Magn√©tique (IRM)

L‚Äôimagerie par r√©sonance magn√©tique (IRM) est une technique d'imagerie m√©dicale non invasive permettant de produire des images anatomiques et fonctionnelles d√©taill√©es du corps humain. Cette technologie repose sur les propri√©t√©s de r√©sonance magn√©tique nucl√©aire (RMN) des atomes d‚Äôhydrog√®ne, tr√®s pr√©sents dans le corps humain. Lorsqu'ils sont expos√©s √† un champ magn√©tique intense (B), les noyaux d‚Äôhydrog√®ne s‚Äôalignent avec ce champ et absorbent des impulsions de radiofr√©quences (RF), modifiant temporairement leur √©tat magn√©tique. Apr√®s l‚Äôarr√™t des impulsions RF, ces noyaux reviennent progressivement √† leur √©tat initial en √©mettant un signal mesurable, utilis√© pour g√©n√©rer des images. Diff√©rentes s√©quences IRM, telles que T Un(a), T Deux (b) et FLAIR(d) , permettent de distinguer les tissus sains des anomalies comme les tumeurs ou l‚Äô≈ìd√®me. Par exemple, les images pond√©r√©es en T Un mettent en avant les structures anatomiques, tandis que les s√©quences FLAIR suppriment le signal du liquide c√©phalorachidien pour mieux visualiser les l√©sions. De plus, l'injection de gadolinium, un produit de contraste, accentue le signal des tumeurs, facilitant ainsi le diagnostic et la planification des traitements. **L‚Äôexemple ci-dessous illustre diff√©rentes s√©quences IRM d‚Äôune tumeur c√©r√©brale** :

# Pr√©traitement des donn√©es

## Lire les donn√©es et les stocker dans un dataframe (df)

```python
# # G√©n√©rer les chemins de donn√©es avec les labels
data_dir = r'.\brain-mri-images-for-brain-tumor-detection' # D√©finir le r√©pertoire contenant les images de MRIs
filepaths = []   # Liste vide pour stocker les chemins des fichiers d'images.
labels = []     # Liste vide pour stocker les Labels correspondant √† chaque image

folds = os.listdir(data_dir)
for fold in folds:
    foldpath = os.path.join(data_dir, fold)
    filelist = os.listdir(foldpath)
    for file in filelist:
        fpath = os.path.join(foldpath, file)
        
        filepaths.append(fpath)
        labels.append(fold)
        
# Concat√©ner les chemins de fichiers avec les labels dans un DataFrame
Fseries = pd.Series(filepaths, name= 'filepaths') # Cr√©er une s√©rie pandas contenant les chemins de fichiers.
Lseries = pd.Series(labels, name='labels')      # Cr√©er une s√©rie pandas contenant les √©tiquettes.
df = pd.concat([Fseries, Lseries], axis= 1)      # Combiner les deux s√©ries en un DataFrame sur l'axe des colonnes.
```

```python
# D√©finir la variable 'strat' qui contient les √©tiquettes
strat = df['labels']  # 'labels' contient les (yes , no) des images.

# Diviser le DataFrame en ensembles d'entra√Ænement et de test
train_df, test_df = train_test_split(df,  # Diviser le DataFrame 'df' en deux sous-ensembles : 'train_df' et 'test_df'.
                                      train_size= 0.8,  # 80% des donn√©es seront utilis√©es pour l'entra√Ænement.
                                      shuffle= True,  # M√©langer les donn√©es avant de les diviser.
                                      random_state= 123,  
                                     # Fixer la graine al√©atoire pour garantir la reproductibilit√© des r√©sultats.
                                      stratify= strat) 
                     # Assurer que la r√©partition des labels dans les ensembles d'entra√Ænement et de test est √©quilibr√©e.
```

## Ici, je continue d'importer les modules n√©cessaires

```python
# Importer les biblioth√®ques n√©cessaires
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential # permet de cr√©er des r√©seaux de neurones couche par couche.
from tensorflow.keras.optimizers import Adam, Adamax # utilis√©s pour ajuster les poids du mod√®le lors de l'entra√Ænement.
from tensorflow.keras.preprocessing.image import ImageDataGenerator # effectuer des augmentations sur les donn√©es d'images.
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Activation, Dropout, BatchNormalization
from tensorflow.keras import regularizers
# Ignorer les avertissements
import warnings
warnings.filterwarnings("ignore") # Ignorer les avertissements pour √©viter leur affichage pendant l'ex√©cution du code.

print ('modules loaded')
```

## Cr√©er un g√©n√©rateur de donn√©es d'image

```python
# Taille des images redimensionn√©es
batch_size = 8  # Le nombre d'images trait√©es par lot (batch) pendant l'entra√Ænement. Ici, 8 images par it√©ration.
img_size = (224, 224)  # La taille de l'image redimensionn√©e, ici 224x224 pixels.
channels = 3  # Nombre de canaux de couleur (3 pour RGB : rouge, vert, bleu).
img_shape = (img_size[0], img_size[1], channels)  # Forme de l'image (224, 224, 3).

# Cr√©ation d'un g√©n√©rateur d'images pour l'entra√Ænement et le test
tr_gen = ImageDataGenerator()  # Cr√©ation d'un g√©n√©rateur pour l'entra√Ænement (sans transformation ici).
ts_gen = ImageDataGenerator()  # Cr√©ation d'un g√©n√©rateur pour le test (sans transformation ici).

# G√©n√©rateur pour les donn√©es d'entra√Ænement
train_gen = tr_gen.flow_from_dataframe(  # Cr√©e un g√©n√©rateur de donn√©es √† partir du DataFrame d'entra√Ænement.
    train_df,  # DataFrame contenant les donn√©es d'entra√Ænement.
    x_col='filepaths',  # Colonne contenant les chemins des fichiers d'images.
    y_col='labels',  # Colonne contenant les √©tiquettes des images (labels).
    target_size=img_size,  # Redimensionnement des images √† 224x224 pixels.
    class_mode='categorical',  # Mode des classes (ici, multi-classes avec √©tiquettes cat√©gorielles).
    color_mode='rgb',         # Mode de couleur (ici, RGB avec 3 canaux).
    shuffle=True,         # M√©langer les donn√©es avant chaque √©poque d'entra√Ænement.
    batch_size=batch_size  # Taille des lots d'images pendant l'entra√Ænement.
)

# G√©n√©rateur pour les donn√©es de test
test_gen = ts_gen.flow_from_dataframe(  # Cr√©e un g√©n√©rateur de donn√©es √† partir du DataFrame de test.
    test_df,  # DataFrame contenant les donn√©es de test.
    x_col='filepaths',  # Colonne contenant les chemins des fichiers d'images.
    y_col='labels',  # Colonne contenant les labels des images .
    target_size=img_size,  # Redimensionnement des images √† 224x224 pixels.
    class_mode='categorical',  # Mode des classes .
    color_mode='rgb',  # Mode de couleur (ici, RGB avec 3 canaux).
    shuffle=False,  # Ne pas m√©langer les donn√©es de test pour pr√©server l'ordre.
    batch_size=batch_size  # Taille des lots d'images pendant l'√©valuation.
)
```

## Afficher 8 images d'entra√Ænement

```python
g_dict = train_gen.class_indices      # D√©finit un dictionnaire {'classe': index} qui mappe chaque classe √† un indice
classes = list(g_dict.keys())         
images, labels = next(train_gen)      # R√©cup√®re un lot d'images et leurs √©tiquettes (labels) √† partir du train generator

plt.figure(figsize=(20, 10))           # Cr√©e une figure avec une taille de 20x10 pouces pour afficher les images

# Boucle pour afficher 10 images (car batch_size=10)
for i in range(8):
    plt.subplot(2, 4, i + 1)          
    image = images[i] / 255            # Redimensionne les pixels de l'image dans la plage [0, 1] pour normaliser les donn√©es
    plt.imshow(image)                  # Affiche l'image dans le sous-graphe actuel
    index = np.argmax(labels[i])       # R√©cup√®re l'indice de la classe √† partir des √©tiquettes (encod√©es en one-hot)
    class_name = classes[index]        # R√©cup√®re le nom de la classe en utilisant l'indice pr√©c√©demment obtenu
    plt.title(class_name, color='blue', fontsize=12)  # Affiche le nom de la classe comme titre de l'image, avec couleur bleue 
    plt.axis('off')                    

plt.show()                             # Affiche toutes les images et leurs titres dans une fen√™tre de visualisation
```

# Structure du mod√®le
## Cr√©ation d'un mod√®le g√©n√©rique

```python
# D√©finition de la taille des images et des param√®tres d'entr√©e
img_size = (224, 224)  # D√©finir la taille des images d'entr√©e (224x224 pixels).
channels = 3  # Nombre de canaux pour l'image (3 pour RGB : Rouge, Vert, Bleu).
img_shape = (img_size[0], img_size[1], channels)  # La forme de l'image pour le mod√®le (224, 224, 3) pour des images RGB.

# D√©finir le nombre de classes bas√© sur les labels des donn√©es d'entra√Ænement
class_count = len(list(train_gen.class_indices.keys()))  # Calcule le nombre de classes.

# Cr√©er le mod√®le de base pr√©-entra√Æn√© EfficientNetB3
base_model = tf.keras.applications.EfficientNetB3(
    include_top=False,  # Exclut la partie "top" (couches finales de classification), pour ajouter notre propre couche de sortie.
    weights="imagenet",  # Utilise les poids pr√©-entra√Æn√©s sur ImageNet.
    input_shape=img_shape,  # D√©finir la forme des images d'entr√©e.
    pooling='max'  # Utilise un pooling de type "max" pour r√©duire la dimensionnalit√© des caract√©ristiques extraites.
)

# Cr√©er le mod√®le s√©quentiel en ajoutant des couches suppl√©mentaires
model = tf.keras.Sequential([
    base_model,  # Ajoute le mod√®le EfficientNetB3 pr√©-entra√Æn√©.
    tf.keras.layers.BatchNormalization(axis=-1, momentum=0.99, epsilon=0.001),  # Normalisation des lots pour am√©liorer la stabilit√©.
    tf.keras.layers.Dense(
        256,  # Ajoute une couche dense avec 256 neurones.
        kernel_regularizer=tf.keras.regularizers.l2(0.016), # la r√©gularisation L2 sur les poids pour √©viter l'overfitting.
        activity_regularizer=tf.keras.regularizers.l1(0.006),  # Applique la r√©gularisation L1 sur les activations.
        bias_regularizer=tf.keras.regularizers.l1(0.006),  # Applique la r√©gularisation L1 sur les biais.
        activation='relu'  # Fonction d'activation ReLU (Rectified Linear Unit) pour introduire de la non-lin√©arit√©.
    ),  # Ajoute une couche Dropout avec un taux de 45% pour pr√©venir le sur-apprentissage.
    tf.keras.layers.Dropout(rate=0.45, seed=123),  
    # Couche de sortie avec activation softmax pour la classification multi-classes.
    tf.keras.layers.Dense(class_count, activation='softmax')  
])
# Compiler le mod√®le avec l'optimiseur et la fonction de perte appropri√©s
model.compile(
    optimizer=tf.keras.optimizers.Adamax(learning_rate=0.001),  # l'optimiseur Adamax avec un taux d'apprentissage de 0.001.
    loss='categorical_crossentropy',  # Fonction de perte pour la classification multi-classes.
    metrics=['accuracy']  # Utilise la pr√©cision comme m√©trique pour √©valuer les performances du mod√®le.
)

# Afficher un r√©sum√© du mod√®le pour visualiser sa structure
model.summary()  # Affiche la structure du mod√®le, les couches, le nombre de param√®tres, etc.
```

## R√©sultats du Mod√®le

Le mod√®le utilise **EfficientNetB3** pr√©-entra√Æn√© comme base, suivi de couches suppl√©mentaires pour la r√©gularisation et la classification. Voici une explication des r√©sultats du mod√®le :

### Couches du mod√®le

- **EfficientNetB3** : Cette couche est responsable de l'extraction des caract√©ristiques √† partir des images d'entr√©e. Elle r√©duit la taille des images tout en conservant les informations importantes pour la classification, produisant une sortie de forme `(None, 1536)`, ce qui signifie que chaque image est repr√©sent√©e par un vecteur de 1536 caract√©ristiques.
  
- **BatchNormalization** : Apr√®s l'extraction des caract√©ristiques, la normalisation par lot est appliqu√©e pour stabiliser l'entra√Ænement en ajustant les activations. Cela aide √† acc√©l√©rer l'apprentissage et √† rendre le mod√®le plus stable.

- **Dense (256 neurones)** : Cette couche enti√®rement connect√©e avec 256 neurones apprend √† combiner les caract√©ristiques extraites pour d√©tecter des patterns complexes dans les donn√©es d'entr√©e. Le nombre de param√®tres associ√©s √† cette couche est de **393,472**.

- **Dropout** : Le taux de **Dropout** de 45 % est appliqu√© pour √©viter le surapprentissage en d√©sactivant al√©atoirement certaines connexions entre les neurones pendant l'entra√Ænement.

- **Dense (sortie)** : Enfin, la couche de sortie avec 2 neurones utilise la fonction **softmax** pour pr√©dire la probabilit√© d'appartenir √† chaque classe (par exemple, pr√©sence ou absence de tumeur). La sortie de cette couche est un vecteur de forme `(None, 2)`.

### Conclusion

Le mod√®le est capable de classifier les images en deux cat√©gories : **tumeur** et **non-tumeur**. La structure du mod√®le, en combinant un r√©seau pr√©-entra√Æn√© avec des techniques de r√©gularisation comme la normalisation par lot et le dropout, permet de capturer des caract√©ristiques complexes tout en √©vitant le surapprentissage. Ce mod√®le devrait √™tre efficace pour la d√©tection de tumeurs sur des images de cerveau, en particulier gr√¢ce √† l'utilisation de **EfficientNetB3**, un mod√®le pr√©-entra√Æn√© performant.


# Entra√Ænement du mod√®le
